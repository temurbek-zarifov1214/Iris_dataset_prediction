```python
import pandas as pd
import numpy as np
```

```python
md=pd.read_csv('Iris.csv')
md
```

```python
md=md.drop('Id',axis=1)
```

```python
md.head()
```

```python
md.columns=["sepal_length","sepal_width","petal_length","petal_width","version"]
```

```python
md.head()
```

```python
x=md.drop("version",axis=1)
```

```python
y=pd.get_dummies(md.version)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
y_train.head()
```

```python
y_train.head()
```

```python
from keras import Sequential
from keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(units=6, activation="sigmoid"))
model.add(Dense(units=3,activation="softmax"))
```

```python
model.compile(optimizer = 'rmsprop' , loss = 'categorical_crossentropy' ,  metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=20, batch_size=5)
```

```python
result=model.evaluate(x_test,y_test)
```
